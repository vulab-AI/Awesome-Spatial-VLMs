```python
import json

def json_to_batch_jsonl(input_json, output_jsonl):
    """
    把包含 benchmark 数据的 JSON 文件转成 Gemini Batch API 的 jsonl
    """
    with open(input_json, "r") as f:
        dataset = json.load(f)

    with open(output_jsonl, "w") as f:
        for item in dataset:
            request_obj = {
                # 直接用原始的 id（字符串或数字都可以）
                "id": str(item["id"]),
                "request": {
                    "contents": [
                        {
                            "role": "user",
                            "parts": [
                                # 图片部分
                                *[
                                    {
                                        "file_data": {
                                            "file_uri": uri,
                                            "mime_type": "image/png"
                                        }
                                    }
                                    for uri in item.get("image_uris", [])
                                ],
                                # 文本 prompt
                                {"text": item["prompt"]+"\nAnswer the question below. If it's selection question, only return the option letter. Else, only return the answer phrase."}
                            ]
                        }
                    ]
                }
            }
            f.write(json.dumps(request_obj, ensure_ascii=False) + "\n")

    print(f"✅ Converted {len(dataset)} records -> {output_jsonl}")
```

```python
import os
files= os.listdir("jsons")
for file in files:
    if file.endswith(".json"):
        json_to_batch_jsonl(os.path.join("jsons", file), os.path.join("jsons", file.replace(".json", ".jsonl")))
```
