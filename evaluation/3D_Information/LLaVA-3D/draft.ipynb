```python
import json
import os
import numpy as np
import re

#load json
folder_path="/home/tuo/Codes/survey_eval/cambrian/output/cambrian"
for file in os.listdir(folder_path):

    path = os.path.join(folder_path, file)
    with open(path, "r") as f:
        data = json.load(f)

    count = 0
    for item in data:
        if "answer" in item["result"].lower():
            #answer and . 之间的内容
            item["result"] = re.split(r'answer[.:]?', item["result"], flags=re.IGNORECASE)[-1].strip()
            if "\n\n" in item["result"]:
                item["result"] = item["result"].split("\n\n")[-1].strip()
                
            if "." in item["result"]:
                item["result"] = item["result"].split(".")[0].strip()

            # print(f" {item['result']}")
        # elif "option" in item["result"].lower():
        #     item["result"] = re.split(r'option[.:]?', item["result"], flags=re.IGNORECASE)[-1].strip()
        #     if "\n\n" in item["result"]:
        #         item["result"] = item["result"].split("\n\n")[-1].strip()
                
            if "." in item["result"]:
                item["result"] = item["result"].split(".")[0].strip()
            # print(f" {item['result']}")
        if 'Whats_Up_results.json' in file:
            print(f" {item['result']}")

        if item['GT'].endswith(";") or item['GT'].endswith("."):
            item['GT'] = item['GT'][:-1].strip()

        if item['GT'].lower().strip() in item['result'].lower().replace(".", " ").strip():
            count += 1
            continue
        elif item["result"].lower().replace(".", " ").strip() in item['GT'].lower().strip():
            count += 1
            continue
    print(f"File: {file}")
    if len(data) == 0:
        accuracy = 0
    else:
        accuracy = count / len(data)
    print(count, len(data))
    print(f"Accuracy: {accuracy*100:.2f}%")
```
