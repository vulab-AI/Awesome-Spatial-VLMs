```python
import json
import os
import numpy as np
import re

#load json
path="/home/tuo/Codes/survey_eval/spaceom_spaceqwen/output/spaceqwen/SEED_Bench_Spatial_results.json"
with open(path, "r") as f:
    data = json.load(f)

count = 0
for item in data:
    if "<answer>" in item["result"]:
        ans = re.split("<answer>", item["result"])[0].strip()
        print(ans)
    else:
        ans = item["result"].strip()
        print(ans)
    if ans is None:
        continue
    if item["GT"]==ans:
        count += 1
accuracy = count / len(data)
print(f"Accuracy: {accuracy*100:.2f}%")
```
